```python
main_page_url = "https://www.bola.net/tag/real-madrid/"  # Replace with the actual main page URL
articles = scrape_all_articles(main_page_url)

# Print extracted articles
for idx, article in enumerate(articles, 1):
    if idx > 3: # For simplicity, we only print the first 3 articles
        break
    print(f"Article {idx}:")
    print(f"Headline: {article['headline']}")
    print(f"Date Published: {article['datePublished']}")
    print(f"Content: {article['articleBody']}\n")
```

```python
df = pd.DataFrame(articles)
df['datePublished'] = pd.to_datetime(df['datePublished'])
```

```python
if df['datePublished'].dtypes == 'datetime64[ns]':
    df = df.sort_values(by='datePublished', ascending=False)
df.head()
```

```python
# Apply the function
df = drop_rows_before_date(df, "2025-01-13")

# Display the updated DataFrame
print(df)
```

```python
# Save the filtered DataFrame to a CSV file
df.to_csv('Berita Elclasico.csv', index=False)
```
